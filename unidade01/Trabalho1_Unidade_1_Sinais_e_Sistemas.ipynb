# Trabalho 1 - Unidade 1
### Objetivo: Rotinas em Python para operações com números complexos (soma subtração, multiplicação e divisão)

```python
# importando bibliotecas
import cmath # CMath -> Ela tem modulo de números imaginários
import math # Math -> permite fazer operações e acessar algumas constantes matemáticas
import random # Permite gerar números aleatório

# Gerando dois números complexos com parte real e imaginária inteira aleatória
z1 = complex(random.randint(-9, 9), random.randint(-9, 9))
z2 = complex(random.randint(-9, 9), random.randint(-9, 9))

print(f'z1 = {z1}')
print(f'z2 = {z2}')
```

# Soma e subtração de números complexos

```python
# -= Soma =-
print('-= Soma =-\n')

# Soma da forma direta
z_sum1 = z1 + z2
print(f'1. {z1} + {z2} = {z_sum1}\n')

# Soma da forma detalhada
z_sum2_real = z1.real + z2.real
z_sum2_imag = z1.imag + z2.imag
z_sum2 = complex(z_sum2_real, z_sum2_imag)

print(f'2. ({z1.real} + {z2.real}) + j({z1.imag} + {z2.imag}) = {z_sum2}\n\n')

# =========================================================================

# -= Subtração =-
print('-= Subtração =-\n')

# Subtração de forma direta
z_sub1 = z1 - z2
print(f'1. {z1} - {z2} = {z_sub1}\n')

# Subtração da forma detalhada
z_sub2_real = z1.real - z2.real
z_sub2_imag = z1.imag - z2.imag
z_sub2 = complex(z_sub2_real, z_sub2_imag)

print(f'2. ({z1.real} - {z2.real}) + j({z1.imag} - {z2.imag}) = {z_sub2}\n\n')
```

# Multiplicação e Divisão
Aqui será abordado duas formas: a forma direta via a as facilidades da linguagem de programação e da forma técnica, no qual vai necessitar transformar para suas forma polar.

```python
# == Transformação para forma polar ==
print("== Transformando para forma polar ==")


print("Forma nativa do CMath:")
z1_polar1 = cmath.polar(z1)
z2_polar1 = cmath.polar(z2)

print(f'1. z1 = {z1} -> {z1_polar1}\n')
print(f'2. z2 = {z2} -> {z2_polar1}\n\n')


print("Forma númerica - Detalhada:")
# modulo (r) -> r = raiz(a^2 + b^2) -> onde: z = a + jb
# fase (fs) -> fs = cotg(b/a)

# Transformando z1 para sua forma polar
z1_modulo = math.sqrt(math.pow(z1.real,2) + math.pow(z1.imag,2))
z1_fase = math.atan2(z1.imag, z1.real) # Calcula o arcotangente dos parametros passado, retornando valores entre [-pi, +pi]
z1_polar2 = (z1_modulo, z1_fase)

# Transformando z2 para sua forma polar
z2_modulo = math.sqrt(math.pow(z2.real,2) + math.pow(z2.imag,2))
z2_fase = math.atan2(z2.imag, z2.real)
z2_polar2 = (z2_modulo, z2_fase)

print(f'1. z1 = {z1} -> {z1_polar2}\n')
print(f'2. z2 = {z2} -> {z2_polar2}\n\n')

# Trnsformando para forma polar de Euler
print("Forma polar de Euler:")
z1_euler = f"{z1_modulo:.2f}e^({z1_fase:.2f})"
z2_euler = f"{z2_modulo:.2f}e^({z2_fase:.2f})"

print(f'1. z1_polar = {z1_polar2} -> {z1_euler}\n')
print(f'2. z2_polar = {z2_polar2} -> {z2_euler}')


```

```python
# Multiplicação
print('== Multiplicação ==\n')

# Forma nativa
print('Forma nativa:')
mult_rect = (z1 * z2)
mult_polar1 = cmath.polar(mult_rect)
mult_euler = f'{mult_polar1[0]:.2f}e^({mult_polar1[1]:.2f})'

print(f'1. {z1} * {z2} = {mult_polar1} -> {mult_euler}\n\n')

# Multiplicação da forma direta
print('Forma direta:')
mult_modulo = (z1_modulo * z2_modulo)
mult_fase = (z1_fase + z2_fase)

mult_polar2 = (mult_modulo, mult_fase)
mult_euler = f'{mult_modulo:.2f}e^({mult_fase:.2f})'

print(f'1. {z1} * {z2} = {mult_polar2} -> {mult_euler}\n')
```

```python
# Divisão
print('== Divisão ==\n')

# Forma nativa
print('Forma nativa:')
div_rect = (z1 / z2)
div_polar1 = cmath.polar(div_rect)
div_euler = f'{div_polar1[0]:.2f}e^({div_polar1[1]:.2f})'

print(f'1. {z1} * {z2} = {div_polar1} -> {div_euler}\n\n')

# Divisão da forma direta
print('Forma direta:')
div_modulo = (z1_modulo / z2_modulo)
div_fase = (z1_fase - z2_fase)
div_fase_normalizada = (div_fase + math.pi) % (2 * math.pi) - math.pi # Faz com que a fase fique dentro do intervalo [-pi, +pi]

div_polar2 = (div_modulo, div_fase_normalizada)
div_euler = f'{div_modulo:.2f}e^({div_fase_normalizada:.2f})'

print(f'1. {z1} * {z2} = {div_polar2} -> {div_euler}\n')
```
